```python
import pandas as pd

data = pd.read_csv('ObesityDataSet.csv')

print(data.head())

print(data.info())

print(data.describe())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Menampilkan distribusi berat badan
plt.figure(figsize=(10, 6))
sns.histplot(data['Weight'], bins=30, kde=True)
plt.title('Distribusi Berat Badan')
plt.xlabel('Berat Badan')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Cek missing values
missing_values = data.isnull().sum()
print("Missing Values:\n", missing_values)

# Cek unique values
unique_values = {col: data[col].nunique() for col in data.columns}
print("Unique Values:\n", unique_values)

# Cek data duplikat
duplicate_rows = data.duplicated().sum()
print("Jumlah Data Duplikat:", duplicate_rows)
```

```python
# Visualisasi untuk mendeteksi outlier menggunakan boxplot
plt.figure(figsize=(10, 3))
sns.boxplot(x='NObeyesdad', y='Weight', data=data)
plt.title('Deteksi Outlier Berat Badan berdasarkan Kategori Obesitas')
plt.show()
```





*  Kesimpulan dari Proses EDA

1.Dataset berisi sejumlah baris dan kolom yang dapat memberikan gambaran tentang data demografis dan kesehatan individu.
2.Distribusi berat badan menunjukkan pola tertentu dan dapat membantu dalam memahami proporsi berat badan di berbagai kategori obesitas.
3.Identifikasi nilai yang hilang dan unik dalam dataset penting untuk pemrosesan data lebih lanjut.
4.Boxplot membantu dalam mendeteksi outlier, yang dapat mempengaruhi analisis lebih lanjut dan perlu ditangani dengan benar.
